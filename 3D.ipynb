```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
n=200

melon_width = np.random.normal(10, 2, n)
melon_length = np.random.normal(20, 4, n)
melon_weight = np.random.normal(3, 1, n)

balloons_width = np.random.normal(5, 2 ,n)
balloons_length = np.random.normal(10, 3, n)
balloons_weight = np.random.normal(0.6, 0.2, n)

width=np.concatenate((melon_width,balloons_width))
length=np.concatenate((melon_length,balloons_length))
weight = np.concatenate((melon_weight,balloons_weight))

x=np.array([width,length,weight]).T

melon_label=np.zeros(n,dtype='int')
balloons_label=np.ones(n,dtype='int')

y=np.concatenate((melon_label,balloons_label))
```
